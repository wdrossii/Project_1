```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib inline
import matplotlib.pyplot as plt
#file_path = file_path = r'InputData/MHRemoteData.csv'
```

```python
df = pd.read_csv("InputData/MHRemoteData.csv")
df.head()
```

```python
df['Physical_Activity'].fillna(0,inplace=True)
df['Physical_Activity'] = df['Physical_Activity'].replace({"Weekly": 1, "Daily": 7, "NaN": 0, "None": 0}, inplace=False)
df.head()
```

```python
bins = [0, 1, 7] 
labels = [ 'Weekly', 'Daily']
df['Physical_Activity'] = pd.cut(df['Physical_Activity'], bins=bins, labels=labels)
df.head()
```

```python
MHRemoteReduced = df[["Employee_ID","Gender", "Age", "Job_Role", "Industry", "Years_of_Experience", "Work_Location", 
                               "Hours_Worked_Per_Week","Work_Life_Balance_Rating","Stress_Level", "Mental_Health_Condition", 
                               "Social_Isolation_Rating", "Satisfaction_with_Remote_Work","Physical_Activity","Sleep_Quality","Region"]]
```

```python
#MHRemoteReduced['Physical_Activity'].fillna(0,inplace=True)
MHRemoteReduced['Physical_Activity'] = df['Physical_Activity'].replace({"Weekly": 1, "Daily": 7, "NaN": 0, "None": 0}, inplace=False)
MHRemoteQ1 = MHRemoteReduced[["Employee_ID", "Work_Location", "Physical_Activity","Region"]]
MHRemoteQ1.head()
```

```python
MHRemoteQ1_grouped = MHRemoteQ1.groupby(['Region']).size()
MHRemoteQ1_grouped
```

```python
# Plot the data
MHRemoteQ1_grouped.plot(kind='bar', figsize=(12, 8))
plt.title('Physical Activity by Region')
plt.xlabel('Frequency of Physical Activity')
plt.ylabel('Person Count')
plt.xticks(rotation=45)
#plt.legend(title='Physical Activity')
plt.show()
```

```python
MHRemoteQ2 = MHRemoteReduced[["Employee_ID", "Work_Location", "Social_Isolation_Rating"]]
```

```python
MHRemoteQ2_grouped = MHRemoteQ2.groupby('Work_Location')['Social_Isolation_Rating'].mean()
MHRemoteQ2_grouped
```

```python
MHRemoteQ2_grouped.plot(kind='bar', figsize=(12, 8))
plt.title('Social Isolation')
plt.xlabel('Work Location')
plt.ylabel('Isolation Rating')
plt.xticks(rotation=45)
#plt.legend(title='Social Isolation by Work Location')
plt.show()
```

```python
MHRemoteQ3 = MHRemoteReduced[["Employee_ID", "Work_Location", "Mental_Health_Condition"]]
```

```python
MHRemoteQ3_Grouped = MHRemoteQ3.groupby(['Mental_Health_Condition', 'Work_Location']).count()
MHRemoteQ3_Grouped
```

```python
# Group by Mental_Health_Condition and Work_Location, and count the number of Employee_IDs in each group - from Tamya's code
MGRemoteQ3Grouped = MHRemoteQ3.groupby(['Mental_Health_Condition', 'Work_Location'], observed=False)['Employee_ID'].count().reset_index()
```

```python
# Pivot for plotting
MGRemoteQ3Grouped_pivot = MGRemoteQ3Grouped.pivot(index='Mental_Health_Condition', columns='Work_Location', values='Employee_ID')
```

```python
# Plot the data
MGRemoteQ3Grouped_pivot.plot(kind='bar', figsize=(12, 8))
plt.title('Mental Health Condition by Work Location')
plt.xlabel('Mental Health Condition')
plt.ylabel('Employee Count')
plt.xticks(rotation=45)
#plt.legend(title='Work Location')
plt.show()
```
