```python
#Import modules.
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#Read file into dataframe.
df = pd.read_csv("C:/Users/melit/OneDrive/Data Analytics/Modules/Module 7 - Project 1/Project 1/Impact_of_Remote_Work_on_Mental_Health.csv")

#Print out first 5 rows.
df.head()
```

```python
#Print out the basic information about the dataframe to identify whether the number of responses is equal or if some responses are missing. Also, to review the data types.
df.info()
```

```python
#Convert string values to integers to be able to calculate.
df['Stress_Level'].replace({'Low': 1, 'Medium': 2, 'High': 3}, inplace=True)
df
```

```python
#Convert yes/no to Boolean. 
df['Access_to_Mental_Health_Resources'].replace({'Yes': True, 'No': False}, inplace=True)
df
```

```python
#Convert string values to integers to be able to calculate.
df['Productivity_Change'].replace({'Decrease': 1, 'No Change': 2, 'Increase': 3}, inplace=True)
df
```

```python
#Convert string values to integers to be able to calculate.
df['Satisfaction_with_Remote_Work'].replace({'Unsatisfied': 1, 'Neutral': 2, 'Satisfied': 3}, inplace=True)
df
```

```python
#Convert string values to integers to be able to calculate.
df['Physical_Activity'].replace({'Daily': 3, 'Weekly': 2}, inplace=True)
df
```

```python
#Enter the word "None" for "NaN" values.
df['Physical_Activity'].fillna('None', inplace=True)
df
```

```python
#Assign an integer value to the string value in order to calculate.
df['Physical_Activity'].replace({'None': 1}, inplace=True)
df
```

```python
#Convert values with a decimal to a whole number in the Physical Activity column.
df['Physical_Activity'] = df["Physical_Activity"].astype(int)
df
```

```python
#Convert string values to integers to be able to calculate.
df['Sleep_Quality'].replace({'Good': 1, 'Average': 2, 'Poor': 3}, inplace=True)
df
```

```python
#Enter the word "Not Reported" for "NaN" values.
df['Mental_Health_Condition'].fillna('Not Reported', inplace=True)
df
```

```python
#Remove Employee_ID column, as there is already an index.
df = df.drop(['Employee_ID'], axis=1)

#Print out first 5 rows to make sure Employee_ID column has been removed from the dataframe.
df.head()
```
